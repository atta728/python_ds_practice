```python
import pandas as pd
data = pd.read_csv("Pokemon.csv")
```

```python
data.head()
```

**Column of High_Low_Speed and High_Low_HP**

```python
mean_hp = data["HP"].mean()
speed_mean = data["Speed"].mean()

def set_hp(a):
  if a < mean_hp:
    return "Low_HP"
  elif a == mean_hp:
    return "Mid_HP"
  else:
    return "High_HP"

def set_speed(value):
  if value < speed_mean:
    return "Low_Speed"
  elif value == speed_mean:
    return "Mid_Speed"
  else:
    return "High_Speed"
```

```python
data["High_Low_HP"] = data["HP"].apply(set_hp)
data["High_Low_Speed"] = data["Speed"].apply(set_speed)
data
```
