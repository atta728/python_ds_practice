```python
import pandas as pd
import numpy as np
data = pd.read_csv('netflix_data.csv')
```

```python
data.head()
```

**adding num_of_cast column**

```python
def set_noc(x):
  if pd.isnull(x):
    return 0
  else:
    return len(x.split(", "))

data["num_of_cast"] = data["cast"].apply(set_noc)
data.tail()
```

**number of tv-shows and movies of each genre**
